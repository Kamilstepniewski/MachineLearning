```python
import numpy as np
import pandas as pd
import random
import matplotlib as plt
from sklearn import datasets,linear_model
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data', sep='\s+' )
df.columns = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX',\
             'PRATIO','B','LSTAT','MDEV']
df.head()
```

```python
df.describe()
```

```python
random.seed(3277)
training,testing = train_test_split(df,test_size = 0.25)
print(training.shape)
print(testing.shape)
```

```python
regr = linear_model.LinearRegression()
training_data = training.drop('MDEV',axis=1)
training_test = training.iloc[:,-1]
print(training_data.shape)
print(training_test.shape)
```

```python
lfit = regr.fit(training_data,training_test)
print('Coefficients: \n',regr.coef_)
```

```python
testing_data = testing.loc[:,['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX',\
             'PRATIO','B','LSTAT']]
testing_test = testing[['MDEV']].values
```

```python
prediction_of_test = regr.predict(testing_data)
```

```python
sum = 0
rows = len(testing_test)
for i in range(rows):
    test = testing_test[i]
    prediction = prediction_of_test[i]
    diff = (test - prediction) ** 2
    sum = sum + diff
mse = sum / rows
print ("MSE " , mse)
```

```python
%matplotlib inline

plt.pyplot.scatter(testing_test, prediction_of_test, color='black')


#Draw lines
x0 = min(testing_test)
x1 = max(testing_test)
y0 = min(prediction_of_test)
y1 = max(prediction_of_test)
plt.pyplot.plot([x0,x1],[y0,y1],color="red")

#Labels
plt.pyplot.xlabel("Actual Price")
plt.pyplot.ylabel("Predicted Price")
plt.pyplot.title("Actual Price and Predicted Price")

plt.pyplot.show()
```
